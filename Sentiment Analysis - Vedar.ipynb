```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
senti = SentimentIntensityAnalyzer()
```

```python
print(senti.polarity_scores("the food is good"))
print(senti.polarity_scores('the food is very very good'))
print(senti.polarity_scores('the food is good but service is bad'))
print(senti.polarity_scores('the food is shit'))
```

```python
#Creating a Function
text_polar = lambda text:senti.polarity_scores(text)['compound']
```

```python
import pandas as pd
df =pd.read_excel(r"C:\Users\Lenovo\Desktop\DATA SETS\Amazon.xlsx")
```

```python
df['Polarity_Score'] =df['reviews.text'].apply(text_polar)
df
```

```python
df.columns
```

```python
#categoriizing thre polarity score
sentiment_mapping = {1:'High_Neg',2:'Neg',3:'Net',4:'Pos',5:'High_Pos'}
map_sentiment = lambda val:np.digitize(val,[-1,-0.8,-0.25,0.25,0.8])
```

```python
df['Sentiment_category'] = df['Polarity_Score'].apply(map_sentiment)
df.head()
```

```python
df['Sentiment_mapping'] = df['Sentiment_category'].map({1:'High_Neg',2:'Neg',3:'Net',4:'Pos',5:'High_Pos'})
df.head()
```

```python
dt = pd.crosstab(index= df['Sentiment_mapping'], columns='count').reset_index()

dt........
```
